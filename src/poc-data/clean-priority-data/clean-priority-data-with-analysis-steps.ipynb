```python
#import 
import pandas as pd
```

```python
#Read raw csv file
data = pd.read_csv("task_priority_data.csv")
data.head(3)
```

```python
# Step 1: Standardizing column names
data.columns = data.columns.str.strip().str.replace('\n', ' ').str.replace('  ', ' ').str.lower().str.replace(' ','_')
```

```python
data.head()
```

```python
# Remove dependency as it has many missing values and not very useful
data = data.drop(columns=['dependency'])
```

```python
data.columns
```

```python
unique_school_year = data['school_year'].unique()
print(unique_school_year)
```

```python
data['school_year'] = data['school_year'].replace({'Master': '6', 'Phd': '7'})
```

```python
unique_school_year = data['school_year'].unique()
print(unique_school_year)
```

```python
# Extract the number of the year only
data['school_year'] = data['school_year'].str.extract('(\d+)')
unique_school_year = data['school_year'].unique()
print(unique_school_year)
```

```python
# Change the datatype into int instead of str
data['school_year'] = pd.to_numeric(data['school_year'], errors='coerce')
unique_school_year = data['school_year'].unique()
print(unique_school_year)
```

```python
data.head()
```

```python
data.to_csv('task_priority_data_half_cleaned.csv', index=False)
```

```python
# Checking datatype for each colunm so we could change it to the disired datatype
for column in data.columns:
    print(column, data[column].dtype)
```

```python
# Change the progress datatype into float64 so that they can enter data like 90.5 
data['uncompleted_task_current_progress_(%)'] = data['uncompleted_task_current_progress_(%)'].astype('float64')
for column in data.columns:
    print(column, data[column].dtype)
```

```python
# Check missing values
missing_values = data.isnull().sum()
missing_values
```

```python
# User more comprehensive column names
column_name_mapping = {
    'group/individual/hybrid': 'task_mode',
    'task_weight_(%)': 'task_weight_percent',
    'time_required_to_finish_the_rest_of_the_work_(h)': 'time_required_hours',
    'uncompleted_task_current_progress_(%)': 'current_progress_percent',
    'uncomplete_task_time_spent_(h)': 'time_spent_hours',
    'priority_(1-5_l-h)': 'priority_level'
}
data.rename(columns=column_name_mapping, inplace=True)
```

```python
data.columns
```

```python
data.head(10)
```

```python
# Replacing '-' with NaN in 'post_date_yyyy/mm/dd'
data['post_date_yyyy/mm/dd'] = data['post_date_yyyy/mm/dd'].replace('-', pd.NA)

# Converting date columns to datetime format
date_columns = ['post_date_yyyy/mm/dd', 'due_date_yyyy/mm/dd', 'current_date_(today)']
for col in date_columns:
    data[col] = pd.to_datetime(data[col], errors='coerce', format='%Y/%m/%d')
data.head(10)
```

```python
for column in data.columns:
    print(column, data[column].dtype)
```

```python
# Fill missing values in difficulty using median as it is categorical data
data['difficulty'] = pd.to_numeric(data['difficulty'], errors='coerce')
data['difficulty'].fillna(data['difficulty'].median(), inplace=True)
data['difficulty'].isnull().sum()
```

```python
missing_values = data.isnull().sum()
missing_values
```

```python
# Remove post_date_yyyy/mm/dd as it has many missing values and not very useful
data = data.drop(columns=['post_date_yyyy/mm/dd'])
```

```python
# User more comprehensive column names
column_name_mapping = {
    'due_date_yyyy/mm/dd': 'due_date',
}
data.rename(columns=column_name_mapping, inplace=True)
```

```python
#Check for Duplicates
duplicates = data.duplicated().sum()
duplicates
```

```python
# Convert column with limited number, discrete set of values to category datatype.
for col in ['school_year', 'task_mode', 'task_type', 'difficulty', 'priority_level']:
    data[col] = data[col].astype('category')
```
